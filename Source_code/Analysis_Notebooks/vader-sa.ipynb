# Importing of Modules

```python
import pandas as pd
import numpy as np
import nltk 
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
nltk.download('vader_lexicon')
```

# Reading of the CSV Fils

```python
df = pd.read_csv("reviewDataClassified.csv")
df.head()
```

# Positive Review Sentimental Analysis

```python
sia = SentimentIntensityAnalyzer()
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row["pos_des_cleaned"]
    myid = row["Unnamed: 0"]
    res[myid] = sia.polarity_scores(text)
```

```python
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns={"index": "Unnamed: 0"})
vaders = vaders.merge(df, how="left", on="Unnamed: 0")
vaders.head()
```

# Negative Review Sentimental Analysis

```python
res2 = {}
for j, row in tqdm(df.iterrows(), total=len(df)):
    text2 = row["neg_des_cleaned"]
    myid2 = row["Unnamed: 0"]
    res2[myid2] = sia.polarity_scores(text2)
```

```python
vaders2 = pd.DataFrame(res2).T
vaders2 = vaders2.reset_index().rename(columns={"index": "Unnamed: 0"})
vaders_final = pd.merge(left=vaders, right=vaders2, on="Unnamed: 0", how="left")
vaders_final.head()
```

# Exports the Final DataFrame as a CSV FIle

```python
vaders_final.to_csv("reviewDataAnalysed.csv", index=False)
```

# Importing of Modules

```python
import pandas as pd
import numpy as np
from nltk.corpus import wordnet
import string
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer
import nltk
import ast
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

# General Data Cleaning

```python
# load data
generaldata = pd.read_csv("generalDataOfficial.csv")

# duplicate room_types column (roomtypes_clean)
generaldata["roomtypes_clean"] = generaldata.loc[:, "room_types"]

# roomtypes_clean, remove S$\xa0, must be string
generaldata["roomtypes_clean"] = generaldata["roomtypes_clean"].str.replace("$", "")
generaldata["roomtypes_clean"] = generaldata["roomtypes_clean"].str.replace("\\", "")
generaldata["roomtypes_clean"] = generaldata["roomtypes_clean"].str.replace("Sxa0", "")


# convert to dictionary
def todict(x):
    selectedhotel = (generaldata["roomtypes_clean"][x])
    newdict = ast.literal_eval(selectedhotel)
    return newdict


# create pricelist for each hotel
def pricelist(x):
    price_list = []

    for prices in x:
        for item in prices:
            price_list.append(item)

    for j in range(0, len(price_list)):
        price_list[j] = price_list[j].replace(",", "")
        price_list[j] = int(price_list[j])

    return price_list


for i in range(len(generaldata["roomtypes_clean"])):
    roomprice = todict(i).values()

    if len(todict(i).values()) > 0:
        generaldata.at[i, "25percentile"] = np.percentile(pricelist(roomprice), 25)
        generaldata.at[i, "50percentile"] = np.percentile(pricelist(roomprice), 50)
        generaldata.at[i, "75percentile"] = np.percentile(pricelist(roomprice), 75)

# remove hotels with no reviews
hotelspacioussuites = generaldata[(generaldata.hotel_id == "spacious-suite-room")].index
hotellings = generaldata[(generaldata.hotel_id == "lings")].index

generaldata.drop(hotelspacioussuites, inplace=True)
generaldata.drop(hotellings, inplace=True)
print("General Data Cleaned")
generaldata.head()
```

# Review Data Cleaning

```python
pd.options.display.max_colwidth = 100
pd.set_option('display.max_rows', None)

# read in the review data
review_data = pd.read_csv("reviewsDataOfficial.csv")
print(len(review_data))

# set seed for reproducibility
np.random.seed(0)


def get_wordnet_pos(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'):
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN


def clean_text(text):
    # lower text
    text = str(text).lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove stop words
    stop = stopwords.words('english')
    text = [x for x in text if x not in stop]
    # remove empty tokens
    text = [t for t in text if len(t) > 0]
    # pos tag text
    pos_tags = pos_tag(text)
    # lemmatize text
    text = [WordNetLemmatizer().lemmatize(t[0], get_wordnet_pos(t[1])) for t in pos_tags]
    # remove words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return text


# split column Date into column Month and column Year
review_data[["month", "year"]] = review_data['date'].str.split(' ', expand=True)

# replace escape character with comma
review_data['positive_des'] = review_data['positive_des'].str.replace("\r", ",")

# remove emojis and special characters from column Title
review_data['title'] = review_data['title'].str.replace(r"[^a-zA-Z!&@, ]+", " ").str.strip()
# print(review_data['title'].dtypes)

# remove emojis and special characters from column positive_des and negative_des
review_data['positive_des'] = review_data['positive_des'].str.replace(r"[^a-zA-Z!&@, ]+", " ").str.strip()
review_data['negative_des'] = review_data['negative_des'].str.replace(r"[^a-zA-Z!&@, ]+", " ").str.strip()

# replace value not in regex with '', and drop rows with nan value
review_data['username'] = review_data['username'].str.replace(r"[^a-zA-Z!&@, ]+", " ").str.strip()
review_data['username'].replace('', np.nan, inplace=True)
review_data.dropna(subset=['username'], inplace=True)

# value count for columns title, positive_des, negative_des
review_data['title'].value_counts()  # the word 'Exceptional' has the most value counts
review_data['positive_des'].mode()  # the word 'Location' has the most value counts
review_data['negative_des'].mode()  # the word 'Nothing' has the most value counts

# clean columns title, positive_des, negative_des
review_data['pos_des_cleaned'] = review_data['positive_des'].apply(lambda x: clean_text(x))
review_data['neg_des_cleaned'] = review_data['negative_des'].apply(lambda x: clean_text(x))
review_data['title_cleaned'] = review_data['title'].apply(lambda x: clean_text(x))

# update values with nil/nan/na/blank
review_data.loc[review_data['pos_des_cleaned'].str.startswith('nil'), 'pos_des_cleaned'] = 'no_pos_des'
review_data.loc[review_data['pos_des_cleaned'].str.startswith('nan'), 'pos_des_cleaned'] = 'no_pos_des'
review_data['pos_des_cleaned'].replace('na', 'no_neg_des', inplace=True)
review_data['pos_des_cleaned'].replace('none', 'no_pos_des', inplace=True)
review_data['pos_des_cleaned'].replace('nothing', 'no_pos_des', inplace=True)
review_data['pos_des_cleaned'].replace('nothinh,', 'no_pos_des', inplace=True)
review_data['pos_des_cleaned'].replace('nope,', 'no_pos_des', inplace=True)

review_data.loc[review_data['neg_des_cleaned'].str.startswith('nil'), 'neg_des_cleaned'] = 'no_neg_des'
review_data.loc[review_data['neg_des_cleaned'].str.startswith('nan'), 'neg_des_cleaned'] = 'no_neg_des'
review_data['neg_des_cleaned'].replace('na', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('na da', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('nathing', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('nik', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('nop', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('nope', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('nothibg', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('nothing', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('non, ', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('none', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('notthing', 'no_neg_des', inplace=True)
review_data['neg_des_cleaned'].replace('notting', 'no_neg_des', inplace=True)

review_data.loc[review_data['title_cleaned'].str.startswith('nil'), 'title_cleaned'] = 'no_title'
review_data.loc[review_data['title_cleaned'].str.startswith('nan'), 'title_cleaned'] = 'no_title'
review_data['title_cleaned'].replace('na', 'no_title', inplace=True)
review_data['title_cleaned'].replace('none', 'no_title', inplace=True)
review_data['title_cleaned'].replace('nothing', 'no_title', inplace=True)
review_data['title_cleaned'].replace('nope', 'no_title', inplace=True)

review_data['pos_des_cleaned'].replace('', np.nan, inplace=True)
review_data['neg_des_cleaned'].replace('', np.nan, inplace=True)
review_data['title_cleaned'].replace('', np.nan, inplace=True)
review_data['pos_des_cleaned'] = review_data['pos_des_cleaned'].replace(np.nan, 'no_pos_des')
review_data['neg_des_cleaned'] = review_data['neg_des_cleaned'].replace(np.nan, 'no_neg_des')
review_data['title_cleaned'] = review_data['title_cleaned'].replace(np.nan, 'no_title')

print("Review Data Cleaned")
review_data.head()
```

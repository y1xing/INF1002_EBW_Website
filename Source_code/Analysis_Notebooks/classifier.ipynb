# Importing of Pandas

```python
import pandas as pd
```

# Classifying of Reviews

```python
# Reads the associative words csv
category_df = pd.read_csv("associative words.csv")
# Stores the column names in a list to help create dicts in the later part of the code
cat_cols = list(category_df.columns)
# The associative words are stored in a dict
cat_dict = {}
for i in cat_cols:
    cat_dict[i] = [x for x in category_df[i].values if str(x) != "nan"]


# Classifies each review intp a category
def get_review_cat(review_string):
    global cat_cols, cat_dict
    # ["No cat"] is returned if there is no review
    if review_string == "no_pos_des" or review_string == "no_neg_des":
        return ["no_cat"]
    # Finds the category of the review
    else:
        # Splits the review string into a list
        str_list = review_string.split(" ")
        review_cat = []
        # Creates a dict to store the score for each category
        similarity_score_dict = {}
        # The scores start at 0
        for j in cat_cols:
            similarity_score_dict[j] = 0
        # Checks if each word in the string matches an associative word
        # If there is a match the score is incremented by
        for j in str_list:
            for key, value in cat_dict.items():
                if j in value:
                    similarity_score_dict[key] += 1
        # Adds categories that match the review into a list
        for key, value in similarity_score_dict.items():
            if value >= 1:
                review_cat.append(key)
        # The review belongs to the General category
        if len(review_cat) == 0:
            review_cat.append("General")
        return review_cat


review_data = pd.read_csv("ReviewDataCleanFinal.csv")

pos_cat = []
neg_cat = []

for i in range(review_data.shape[0]):
    pos_cat.append(get_review_cat(review_data["pos_des_cleaned"][i]))
    neg_cat.append(get_review_cat(review_data["neg_des_cleaned"][i]))

# Adds the categories to the review_data dataframe
review_data["pos_cat"] = pos_cat
review_data["neg_cat"] = neg_cat
review_data.head()
```

# Export the DataFrame to a CSV file

```python
review_data.to_csv("reviewDataClassified.csv")
```
